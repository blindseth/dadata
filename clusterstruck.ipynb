### Clustering using iris data ---------------------------------------------------

```python
import sklearn.datasets as datasets
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import pandas as pd
```

```python
iris1=datasets.load_iris()
df_iris=pd.DataFrame(iris1.data, columns=iris1.feature_names)
y=iris1.target
```

```python
kme = KMeans(init='k-means++',n_clusters=3,n_init=10)
kme.fit(df_iris)

# k_means = KMeans(init='k-means++', n_clusters=3, n_init=10)
# k_means.fit(X)
# print(kme)
```
